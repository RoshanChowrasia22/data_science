```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as ss

df=pd.read_csv("/content/All_India_Index_july2019_20Aug2020_dec20_1_2.csv")
pp=df[['Month','Cereals and products']]
July=df[df['Month']=='July']#Taking data set for month of July only
July_RuralUrban=July[July['Sector']=='Rural+Urban']# Filtering the data set based on Sector
#July_ss=July[['Sector','Cereals and products']]
#July_RuralUrban

###Line plot for Rate of Cereals and products And Vegetables in Rural+Urban along the different years for Month of July
ruy=np.array(July_RuralUrban['Year'])
cp=np.array(July_RuralUrban['Cereals and products'])
veg=np.array(July_RuralUrban['Vegetables'])
fig,at = plt.subplots(figsize=(10,4))
plt.plot(ruy,cp,c='orange',label='Cost of Cereals and products in INR over the years in Month of July',marker='o')
plt.plot(ruy,veg,c='red',label='Cost of Vegetables in INR over the years in Month of July',marker='o')
for index in range(len(cp)):
  at.text(ruy[index],cp[index],(ruy[index],cp[index]))
  at.text(ruy[index],veg[index],(ruy[index],veg[index]))
plt.grid()
plt.xlabel('Over the years in Rural + Urban')
plt.ylabel('Cost in INR')
plt.legend()
plt.show()

##Scatter Plot
August=df[df['Month']=='August']#Filtering data for month of August,
August_Urban=August[August['Sector']=='Urban']#in the Urban region
yr=np.array(August_Urban['Year'])
fuel=np.array(August_Urban['Fuel and light'])
he=np.array(August_Urban['Health'])
tr=np.array(August_Urban['Transport and communication'])
fig,at = plt.subplots(figsize=(10,4))
plt.scatter(yr,fuel,c='orange',label='Change over years in prcies of Fuel and Light in Month of August',marker='o')
plt.scatter(yr,he,c='red',label='Change over years in prcies of Health Care in Month of August',marker='>')
plt.scatter(yr,tr,c='green',label='Change over years in prcies of Transport and Communication in Month of August',marker='<')
for index in range(len(yr)):
  at.text(yr[index],fuel[index],(yr[index],fuel[index]))
  at.text(yr[index],he[index],(yr[index],he[index]))
  at.text(yr[index],tr[index],(yr[index],tr[index]))
plt.grid()
plt.xlabel('Over the years in Urban')
plt.ylabel('Cost in INR')
plt.legend()
plt.show()


###Bar Plot
fig,at = plt.subplots(figsize=(10,4))
pti=np.array(August_Urban['Pan, tobacco and intoxicants'])
plt.bar(yr,pti,label='Change over years in prcies of Pan, tobacco and intoxicants in Month of August',color="orange")
plt.xlabel('Over the years in Urban')
plt.ylabel('Cost in INR')
plt.plot(yr,tr,c='black',label='Change over years in prcies of Transport')

for index in range(len(yr)):
  at.text(yr[index],pti[index],(yr[index],pti[index]),c="blue")
  at.text(yr[index],tr[index],(yr[index],tr[index]),c="black")
plt.grid()
plt.legend()
plt.show()

###violin plot
fig,at = plt.subplots(figsize=(14,14))
ss.violinplot(data=df,x="Sector",y='Meat and fish',hue="Year")
plt.xlabel("Distribution of Prices of Meat and Fish in INR")
plt.grid()
plt.show()
####BOX PLOT
plt.title("The distribution of prices of Cereals and Products")
df[['Cereals and products']].boxplot(grid=True)

#plt.boxplot(df[['Cereals and products']])
###Pair Plot
# fuel_light=np.array(df['Fuel and light'])
# transport_communication=np.array(df['Transport and communication'])
# dataforpairplot=pd.DataFrame()
# dataforpairplot['Fuel and light']=fuel_light
# dataforpairplot['Transport and communication']=transport_communication
# ss.pairplot(dataforpairplot)
# plt.show()
```
